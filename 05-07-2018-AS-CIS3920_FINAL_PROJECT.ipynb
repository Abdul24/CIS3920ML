```python
import pandas as pd 
import numpy as np
import matplotlib as plt 
import seaborn as sns 
import re
```

```python
df.drop(["Minimum Qual Requirements","Preferred Skills","Job ID","To Apply","Work Location",
         "Hours/Shift","Work Location 1","Post Until","# Of Positions","Title Code No",                
"Additional Information", "Posting Updated", "Process Date","Recruitment Contact","Agencies by Topic","New Job Categories"], axis=1, inplace = True)
```

```python
df.columns
```

```python
pd.isnull(df).sum()
```

```python
df["Full-Time/Part-Time indicator"] = df["Full-Time/Part-Time indicator"].astype(object).fillna("F")
df["Master's Degree"] = df["Master's Degree"].astype(object).fillna("No")
df["Bachelor's Degree"] = df["Bachelor's Degree"].astype(object).fillna("No")
df["Microsoft"] = df["Microsoft"].astype(object).fillna("No")
df["Programming experience"] = df["Programming experience"].astype(object).fillna("No")
df["Foreign Language"] = df["Foreign Language"].astype(object).fillna("No")
```

```python
df[['Salary']] = df[['Salary']].fillna(value=76000)
df[['Salary Buckets']] = df[['Salary Buckets']].fillna(value=4)
```

```python
pd.isnull(df).sum()
```

```python
null_data = df[df.isnull().any(axis=1)]
```

```python
null_data
```

```python
df = df.drop(df.index[[1898, 1909, 1910]])
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.replace('\s+', '_')
```

```python
df.head(1)
```

```python
df['Posting_Date'] = pd.to_datetime(df.Posting_Date)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["year"]= df['Posting_Date'].dt.year
df["month"]= df['Posting_Date'].dt.month
df["day"]= df['Posting_Date'].dt.day
```
